```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from pywander.utils.plot_utils import draw, Points, Segments, Polygon, Arrow, Grid
```

```python
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]

_, ax = plt.subplots()

draw(
    ax,
    Points(*dino_vectors)
)
```

```python
_, ax = plt.subplots()
draw(
    ax,
    Points(*dino_vectors),
    Segments((6,4),(3,1))
)
```

```python
_, ax = plt.subplots()
draw(
    ax,
    Grid((-2,-2), (2,2), x_ticks=1, y_ticks=1),
    Arrow((0,0),(2,-2))
)
```

```python
_, ax = plt.subplots()
draw(
    ax,
    Points(*dino_vectors),
    Polygon(*dino_vectors)
)
```
