### 如何引入matplotlib
下面的引用语句已经是惯例了。 

第二句是让图片直接在ipynb上显示，不需要调用 `plt.show()` 了。

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### 连线图
输入数据描述x和y，marker可以指定数据点的形状，各个数据点之间有直线连接起来。

```python
from pywander.utils.plot_utils import line_plot
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
_, (ax1, ax2) = plt.subplots(1, 2, )
line_plot(ax1, data1, data2, marker='x')
line_plot(ax2, data3, data4, marker='o')
```

#### 一个简单的绘制真实值和预测值的例子

```python
xdata = [1,2,3,4,5]
ydata = [1.1,2.1,2.9,3.9,5.1]
predict_ydata = [1,2,3,4,5]
_, ax = plt.subplots()
line_plot(ax, xdata, ydata, color='blue')
line_plot(ax, xdata, predict_ydata, color='red', linestyle='dashed')
```

#### 一个简单的带标签的折线图

```python
y = [2,2.1, 2.2, 2.1, 2.3, 2.4]
_, ax = plt.subplots()
line_plot(ax, y_values=y, x_label='month', y_label='profit', y_lim=(0,2.8), x_tick_labels=['Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug'], marker='o')
```
